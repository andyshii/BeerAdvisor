```python
%%time
chrome_options = Options()
chrome_options.add_argument("--headless")
if first_index == 0:
    print('start...')
else:
    print('continue...')

for i in range(first_index, len(data)):
    
    print(f'site #{i} из {len(data) -1} - {data["Название"][i]}')

    if data['Ссылка_на_Untappd'][i] == '-':
        print('__________________Page = "-"')
        continue
    else:
        site = data['Ссылка_на_Untappd'][i]
        driver = webdriver.Chrome(options=chrome_options)
        driver.get(site)

        sleep(2)

        soup = BeautifulSoup(driver.page_source, 'html.parser')


        name.append(data['Название'][i])
        sort_pv.append(soup.find_all('p', class_='style')[0].get_text())
        abv.append(soup.find_all('p', class_='abv')[0].get_text().split()[0])
        ibu.append(soup.find_all('p', class_='ibu')[0].get_text().split()[0])
        raitings.append(soup.find_all('span', class_='num')[0].get_text()[1:-1])
        check_text = soup.find_all('div', class_='beer-descrption-read-less')[0].get_text()[1:-11]
        if len(check_text) > 0:
            text.append(check_text)
        else: text.append('Описание отсутствует')


        if (i % 10 == 0 and i > 0) or (i == (len(data)-1)) :
            new_data = pd.DataFrame(data={'Название пива': name,
                             'Сорт пива': sort_pv,
                             'ABV': abv,
                             'IBU': ibu,
                             'Оценка': raitings,
                             'Описание': text})

            new_data.to_csv(f'~/Рабочий стол/datasets_pivo/dataset_path_{num_names}.csv', sep=',', index=False)
            num_names += 1
            print('save success')

            name = []
            sort_pv = []
            abv = []
            ibu = []
            raitings = []
            text = []
    
    first_index +=1

    
print('success')
```
